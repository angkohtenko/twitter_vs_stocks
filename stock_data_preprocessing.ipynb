```python
import requests
import pandas as pd
from yahoo_fin.stock_info import get_data
```

```python
# historical daily data from Yahoo finance
tesla_df = get_data("tsla", start_date = None, end_date = None, index_as_date = False, interval="1d")
tesla_df
```

```python
# Drop the NaNs.
tesla_df.dropna()
```

```python
# Drop the ticker column
del tesla_df['ticker']
tesla_df.head(10)
```

```python
# Calculate change in stock price
tesla_df['change'] = tesla_df['close'].diff()
tesla_df.head(10)
```

```python
# Determine data types for each column
tesla_df.dtypes
```

```python
tesla_df.to_csv('Data/tesla_stocks.csv', index=False)
```
